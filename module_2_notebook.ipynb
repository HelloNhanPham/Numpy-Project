```python
x = np.array([4, 2, 1, 2])
y = np.array([1, 2, 2, 0])

def compute_cosine_similarity(x: np.ndarray, y: np.ndarray) -> np.ndarray:
    return np.dot(x, y) / np.multiply(np.linalg.norm(x), np.linalg.norm(y))

np.round(compute_cosine_similarity(x, y), 2)
```

```python
alpha = 5

np.round(compute_cosine_similarity(x*alpha, y), 2)
```

```python
alpha = 0

np.round(compute_cosine_similarity(x*alpha, y), 2)
```

```python
from PIL import Image

image1 = Image.open('./images/lion1.jpg')
image2 = Image.open('./images/dog1.jpg')
query_image = Image.open('./images/lion2.jpg')
```

```python
image1
```

```python
image1.size, image2.size, query_image.size
```

```python
image2 = image2.resize((275, 183))

image2
```

```python
img1_data = np.array(image1)
img2_data = np.array(image2)
query_img_data = np.array(query_image)

img1_data.shape, img2_data.shape, query_img_data.shape
```

```python
img1_gray = np.sum(img1_data, axis=-1) / 3
img2_gray = np.sum(img2_data, axis=-1) / 3
query_img_gray = np.sum(query_img_data, axis=-1) / 3
```

```python
img1_flatten = img1_gray.flatten()
img2_flatten = img2_gray.flatten()
query_img_flatten = query_img_gray.flatten()
```

```python
def compute_cosine_similarity(A, B):
    cosine = np.dot(A, B) / (np.linalg.norm(A) * np.linalg.norm(B))
    return cosine
```

```python
compute_cosine_similarity(img1_flatten, query_img_flatten)
```

```python
compute_cosine_similarity(img2_flatten, query_img_flatten)
```

```python
def compute_cosine_similarity(list_vector, query_vector):
    return list_vector @ query_vector / (np.linalg.norm(list_vector, axis=1) * np.linalg.norm(query_vector))
```

```python
dataset_img_flatten = [img1_flatten, img2_flatten, query_img_flatten]
list_results = compute_cosine_similarity(np.array(dataset_img_flatten), query_img_flatten)
list_results
```
