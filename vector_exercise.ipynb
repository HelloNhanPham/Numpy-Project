### Exercise 1: Compute the length of vector

```python
def compute_vector_length_1(vector):
    return np.sqrt(np.sum(np.power(vector, 2)))
```

```python
def compute_vector_length_2(vector):
    return np.sqrt(np.sum(vector*vector))
```

```python
def compute_vector_length_3(vector):
    return np.linalg.norm(vector)
```

```python
arr = np.arange(5)

print('Solution 1: ',compute_vector_length_1(arr))

print('Solution 2: ',compute_vector_length_2(arr))

print('Solution 3: ',compute_vector_length_3(arr))
```

### Exercise 2: Compute dot product of 2 vectors

```python
def compute_dot_product(vector_1: np.ndarray, vector_2: np.ndarray) -> int:
    return np.sum(vector_1 * vector_2)
```

```python
a = np.array([1, 2, 3])
b = np.array([3, 2, 1])

compute_dot_product(a, b)
```

```python
np.dot(a, b)
```

### Exercise 3: Multiply matrix with vector

```python
def multiplyMatrixWithVector(matrix: np.ndarray, vector: np.ndarray) -> np.ndarray:
    return np.dot(matrix, vector)
```

```python
matrix = np.eye(2, dtype=int)
vector = np.full(shape=(2), fill_value=1)

print('Matrix:\n', matrix)
print('Vector:\n', vector)
```

```python
multiplyMatrixWithVector(matrix, vector)
```

```python
matrix @ vector
```

```python
import cv2
import matplotlib.pyplot as plt
```

```python
def showImage(img):
    plt.imshow(img, cmap=None)
    plt.axis('off')
    plt.show()
```

```python
img = cv2.imread('1.jpg')
img = cv2.resize(src=img, dsize=(100, 100))
img = img[:,:,::-1]

showImage(img)
```

```python
x_0, y_0, x_1, y_1 = 22, 20, 78, 99

red_color = np.array([255, 0, 0])

img[y_0,x_0:x_1] = red_color
img[y_1,x_0:x_1] = red_color
img[y_0:y_1,x_0] = red_color
img[y_0:y_1,x_1] = red_color

showImage(img)
```

```python
cropped_image = img[y_0+1:y_1, x_0+1:x_1, :]
showImage(cropped_image)
```

```python
plt.imsave('output/cropped_image.jpg', cropped_image)
cropped_image = cropped_image[:,:,::-1]
cv2.imwrite('output/cropped_image.png', img=cropped_image)
```
