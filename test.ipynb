```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def showImage(image):
    plt.imshow(image)
    plt.axis('off')
    plt.show()

def convertBGR2RGB(image):
    return image[:,:,::-1]
```

```python
img = cv2.imread('1.jpg')
```

```python
img
```

```python
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

showImage(img)
```

```python
img = cv2.resize(img, (300, 300))

showImage(img)
```

```python
img = img.astype(float)

img = img - 100

img = np.clip(img, a_min=0, a_max=255).astype(np.uint8)

showImage(img)
```

```python
img = convertBGR2RGB(img)

showImage(img)
```

```python
def calculateStableSoftmax(X):
    exps = np.exp(X - np.max(X))
    return exps / np.sum(exps)
```

```python
X = [1.0, 2.0, 3.0]

calculateStableSoftmax(X)
```
