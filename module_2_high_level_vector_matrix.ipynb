### Einstein Summation

```python
u = np.array([1, 2, 3])
v = np.array([1, 2, 3])

r = np.einsum('i, i->', u, v)
r
```

```python
u = np.arange(9).reshape((3, 3))
np.einsum('ii->i', u)
```

```python
u = np.arange(9).reshape((3, 3))
np.einsum('ii->', u)
```

```python
np.einsum('ij->', u)
```

```python
np.einsum('jj->j', u)
```

```python
a = np.arange(6).reshape((2, 3))
b = np.arange(6).reshape((2, 3))

np.einsum('ij,ij->ij', a, b)
```

```python
a = np.arange(6).reshape((2, 3))
b = np.arange(6).reshape((3, 2))

np.einsum('ij, jk -> ik', a, b)
```

```python
a = np.arange(3)
b = np.arange(3)

np.einsum('i,j->ij', a, b)
```
